```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('COVID1')  
df
```

```python
y = df[df['countriesAndTerritories']=='Afghanistan']['cases'][:120]
days = df[df['countriesAndTerritories']=='Afghanistan']['cases'].count()
X = list(range(120,0,-1))
plt.plot(X,y, color='red')
#plt.title('Stock Index Price Vs Time', fontsize=14)
plt.xlabel('Days', fontsize=14)
plt.ylabel('Cases', fontsize=14)
plt.grid(True)
plt.show()
y
```
